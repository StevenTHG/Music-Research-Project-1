```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
music_df1 = pd.read_csv(
    '../Music-Research-Project-1/Resources/top10s.csv', encoding='ISO-8859-1')
music_df1.head(10)
```

```python

music_genre_df1 = music_df1['top genre'].unique()
music_genre_df1
```

```python
music_df1['top genre'] = music_df1['top genre'].replace(['dance pop', 'canadian pop', 'barbadian pop',
                                                       'australian pop', 'indie pop', 'art pop', 'colombian pop',
                                                      'acoustic pop', 'baroque pop', 'candy pop', 'folk-pop', 'colombian pop',
                                                       'danish pop', 'moroccan pop',
                                                       'boy band', 'permanent wave', 'metropopolis', 'escape room',
                                                       'french indie pop'], 'pop')

music_df1['top genre'] = music_df1['top genre'].replace(['detroit hip hop', 'atl hip hop', 'chicago rap',
                                                      'canadian hip hop', 'australian hip hop', 'hip hop', 'hip pop'], 'hip hop')

music_df1['top genre'] = music_df1['top genre'].replace(['big room', 'house', 'australian dance', 'tropical house',
                                                      'belgian edm', 'electronic trap', 'electro house',
                                                       'brostep', 'electropop', 'escape room', 'electro',
                                                       'complextro', 'downtempo'], 'edm')

music_df1['top genre'] = music_df1['top genre'].replace(['neo mellow', 'alaska indie', 'celtic rock', 'hollywood', 'latin', 'canadian latin',
                                                      'contemporary country', 'irish singer-songwriter',
                                                       'british soul'], 'rock/country/other')

music_df1['top genre'] = music_df1['top genre'].replace(
    ['canadian contemporary r&b', 'alternative r&b'], 'R&B')
music_df1
```

```python
music_genre_df1 = music_df1['top genre'].unique()
music_genre_df1
```

```python
pop_df1 = music_df1[['top genre', 'year']].value_counts()
pop_df1.head(10)
```

```python
pop_2011_df = music_df1.loc[(music_df1['top genre'] == 'pop') & (music_df1['year'] == 2011)]
pop_2011_df
```
