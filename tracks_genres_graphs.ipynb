```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
music_df = pd.read_csv('../Music-Research-Project-1/Resources/top10s.csv', encoding = 'ISO-8859-1')
music_df.head(10)
```

```python
music_genre_df = music_df['top genre'].unique()
music_genre_df
```

```python
music_df['top genre'] = music_df['top genre'].replace(['dance pop','canadian pop','barbadian pop',
                                                       'australian pop','indie pop','art pop','colombian pop',
                                                      'acoustic pop','baroque pop','candy pop','folk-pop','colombian pop',
                                                      'danish pop','moroccan pop',
                                                       'boy band','permanent wave','metropopolis','escape room',
                                                      'french indie pop'],'pop')

music_df['top genre'] = music_df['top genre'].replace(['detroit hip hop','atl hip hop','chicago rap',
                                                      'canadian hip hop','australian hip hop','hip hop','hip pop'],'hip hop')

music_df['top genre'] = music_df['top genre'].replace(['big room','house','australian dance','tropical house',
                                                      'belgian edm','electronic trap','electro house',
                                                       'brostep','electropop','escape room','electro',
                                                       'complextro','downtempo'],'edm')

music_df['top genre'] = music_df['top genre'].replace(['neo mellow','alaska indie','celtic rock','hollywood','latin','canadian latin',
                                                      'contemporary country','irish singer-songwriter',
                                                      'british soul'],'rock/country/other')
                                                      
music_df['top genre'] = music_df['top genre'].replace(['canadian contemporary r&b','alternative r&b'],'R&B')
music_df
```

```python
music_genre_df = music_df['top genre'].unique()
music_genre_df
```

```python
pop_df = music_df['top genre'].value_counts()
pop_df
```

```python
genre1_df = music_df[music_df['top genre'] == 'rock/country/other']
genre1_df
```

```python
year_spch = music_df.groupby(['year'])['spch'].mean()
year_spch_plot = year_spch.plot(kind = 'bar', x = 'year', y= 'spch', xlabel = 'Year',ylabel ='Average Speech Level' )
```

```python
year_dB = music_df.groupby(['year'])['dB'].mean()
year_spch_plot = year_dB.plot(kind = 'bar', x = 'year', y= 'dB', xlabel = 'Year',ylabel ='Average Decibel Level' )
```

```python
year_live = music_df.groupby(['year'])['live'].mean()
year_live_plot = year_live.plot(kind = 'bar', x = 'year', y= 'live', xlabel = 'Year',ylabel ='Average Liveliness' )
```

```python
year_val = music_df.groupby(['year'])['val'].mean()
year_val_plot = year_val.plot(kind = 'bar', x = 'year', y= 'val', xlabel = 'Year',ylabel ='Average Valence Level' )
```

```python
year_acous = music_df.groupby(['year'])['acous'].mean()
year_acous_plot = year_acous.plot(kind = 'bar', x = 'year', y= 'acous', xlabel = 'Year',ylabel ='Average Acoustic Level' )
```

```python
import numpy as np
music_df_slim = music_df[['year','top genre','title']]
year_genre = pd.pivot_table(music_df_slim,values="title",index="year", columns="top genre", aggfunc=np.count_nonzero)
# year_genreT = year_genre.set_index('year').T
#index = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
#x_axis = 'year'
#line_graph = year_genre.plot.line()
# plt.plot('year' == 2010, '')
year_genre.plot(kind='line', title="Number of Songs in Top Ten by Genre", ylabel="Number of Songs", xlabel="Year").legend(loc='best')

```
