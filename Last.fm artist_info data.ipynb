```python
import requests 
import json
import pandas as pd
response = requests.get("http://ws.audioscrobbler.com/2.0/?method=chart.gettopartists&api_key=5b02f925244097e08aac37907f8e15ff&format=json").json()
print(response['artists']['artist'])
artist_info = response['artists']['artist']
    
```

```python
df = pd.DataFrame(artist_info)
artist_df = df[['name','playcount','listeners']]
artist_df
```

```python
ordered_playcount_df = artist_df.sort_values(by ='playcount',ascending= True)
ordered_playcount_df
```

```python
ordered_listeners_df = artist_df.sort_values(by = 'listeners', ascending= True)
ordered_listeners_df
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

x_axis = ordered_playcount_df ['name']
y_axis = ordered_playcount_df ['playcount']
plt.figure(figsize = (10,17))
plt.bar(x_axis, y_axis, align="center")
plt.xticks(rotation = 90)
plt.xlabel('Artists')
plt.ylabel('Playcount')
plt.tight_layout
plt.show()
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

x_axis = ordered_listeners_df['name']
y_axis = ordered_listeners_df['listeners']
plt.figure(figsize = (10,17))
plt.bar(x_axis, y_axis, align="center")
plt.xticks(rotation = 90)
plt.xlabel('Artists')
plt.ylabel('Listeners')
plt.tight_layout
plt.show()
```
